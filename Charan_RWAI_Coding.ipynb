```python
import pandas as pd
df1 = pd.read_csv("books.csv")
df1
```

```python
df2 = pd.read_csv("book_tags.csv")
df2
```

```python
df3 = pd.read_csv("ratings.csv")
df3
```

```python
null_values = df1['original_title'].isnull().sum()
null_values
```

```python
df1 = df1.dropna(subset=['original_title'])

x = df1['book_id'].tolist()
df2 = df2[df2['goodreads_book_id'].isin(x)]
df3 = df3[df3['book_id'].isin(x)]


unique = df1['book_id'].nunique()
unique
```

```python
df3['user_id'].nunique()
```

```python
df2['tag_id'].nunique()
```

```python
cnts = df2['tag_id'].value_counts()
cnts.idxmax()
```

```python
merge = pd.merge(df2, df1, left_on='goodreads_book_id', right_on='book_id')

result = merge.groupby('title')['tag_id'].count()

result.idxmax()

```

```python
import matplotlib.pyplot as plt
tag = df2.groupby('tag_id')['count'].sum()

sorted_tags = tag.sort_values(ascending=False)

top_20= sorted_tags.head(20)

plt.figure(figsize=(20, 30))
top_20.plot(kind='bar')
plt.title('Top 20 Unique Tags')
plt.xlabel('Tag ID')
plt.ylabel('Number of User Records')
plt.xticks(rotation=45, ha='right')
plt.show()
```
